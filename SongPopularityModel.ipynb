For this model the data needs to be 
- processed to make sure all nulls and empty rows/columns and taken care of 
- maybe create a pipeline so that we can process the data 
- split in a 70 train / 30 test so that we make sure we can validate if our model is accurate 
- validated after to see how accurate or close our model is in predicting if a song is popular or not


```python
data = data.drop(['Unnamed: 0', 'Track.Name'], axis=1)
```

```python
data
```

```python
data['Popularity'] = pd.qcut(data['Popularity'], q=2, labels=[0,1])

# This means that we want to group songs based on their popularity scoring by giving them a 0 or a 1 on whether they are popular or not 
```

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(data, 'Genre', 'genre')
data = onehot_encode(data, 'Artist.Name', 'artist')
```

```python
data
```

Now we want to split the data into the thing that we want to predict. popularity and all of the quantitive things that could help us. 

```python
Y = data.loc[:,'Popularity']
X = data.drop('Popularity', axis = 1)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Now we want to split the data between the testing and training standard 70 - 30 split 

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.30, random_state=20)
```

```python
log_model = LogisticRegression()
# This will fit the model according the training data split that we gave it 
log_model.fit(X_train,y_train)
```

```python
log_model_acc = log_model.score(X_test, y_test)
print(log_model_acc)
```
